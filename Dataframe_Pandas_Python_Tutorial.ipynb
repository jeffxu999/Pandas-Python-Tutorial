```
# Dataframe - Pandas Python Tutorial

######################################################

import pandas as pd
import random

ids = [i+200 for i in range(100)]

salaries = [random.randint(2500,3500) for i in range(100)]

some_data = {"ID":ids,"Salary":salaries}

print(some_data["ID"])

print(some_data["Salary"])

df = pd.DataFrame(some_data)

print(df)

######################################################

print(df.head())

print(df.head(1))

print(df.tail())

print(df.tail(1))

print(df["ID"])

print(df["Salary"])

######################################################

import matplotlib.pyplot as plt

df.plot()

df["ID"].plot()

df["Salary"].plot()

plt.show()
```
