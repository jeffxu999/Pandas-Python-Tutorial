```
# Merge, Join, & Concatenate - Pandas Python Tutorial

######################################################

import pandas as pd

some_list = [i*2 for i in range(10)]
some_other_list = [i*5 for i in range(10)]

some_df = pd.DataFrame(some_list, columns=["Some DF"])

some_other_df = pd.DataFrame(some_other_list,columns=["Some Other DF"])

print(some_df)
print(some_other_df)

master_df = pd.concat([some_df, some_other_df],axis=1)

print(master_df)

print(some_df)
print(some_other_df)

######################################################

key = [i for i in range(10)]
other_key = [i+10 for i in range(10)]

some_df = pd.DataFrame({"Key":key,"Some DF":some_list})

some_other_df = pd.DataFrame({"Key":key,"Some Other DF":some_other_list})

print(some_df)
print(some_other_df)

master_df = pd.merge(some_df,some_other_df,on="Key")

print(master_df)

print(some_df)
print(some_other_df)

######################################################

master_df = some_other_df.set_index("Key").join(some_df.set_index("Key"))

print(master_df)

print(some_df)
print(some_other_df)
```
